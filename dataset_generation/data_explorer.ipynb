```python
import pandas as pd
base = 'data/'
files = ['bbc.txt', 'cnn.txt', 'fox.txt', 'ndtv.txt', 'nyt.txt', 'wsj.txt']
df = pd.DataFrame()
# df = pd.read_csv(base+files[0])
# source = files[0][:files[0].index('.')]
for file in files:
    source = file[:file.index('.')]
    print(source)
    temp = pd.read_csv(base+file)
    temp['source'] = source
    if source == 'fox':
        temp = temp.sample(frac=.6)
    df = df.append(temp)
print(len(df))
```

```python
df['source'].value_counts().plot(kind='bar', title='Number of Datapoints by Source')
```

```python
print("TOTAL: ", len(df))
print("TOTAL POSITIVE: ", len(df[df.is_article>0]))
print("% POSITIVE: ", len(df[df.is_article>0])/len(df))
df['is_article'].value_counts().plot(kind='bar', title='Number of Datapoints by Class')
```

```python
sources = df.source.unique()
for source in sources:
    print("SOURCE: ", source)
    print("TOTAL: ", len(df[df.source == source]))
    print("TOTAL POSITIVE: ", len(df[(df.is_article>0) & (df.source == source)]))
    print("% POSITIVE: ", len(df[(df.is_article>0) & (df.source == source)])/len(df[df.source == source]))
    print("=========================================")
```
