```python
import sys
sys.path.append("../modules/")
import scraper
from bs4 import BeautifulSoup

html = scraper.fetch_html("https://www.bbc.com/news", waitUntil='load')
soup = BeautifulSoup(html, 'html.parser')
soup = scraper.clean_html(soup)
soup = scraper.squash_nested_divs(soup)
tags = scraper.extract_tags(soup)

for i, tag in enumerate(tags):
    print(f'{i}: {tag}\n')
```

```python
import pandas as pd

df = pd.DataFrame(tags)
df.columns = ['html']
df['is_article'] = 0
df['i'] = df.index
df.head(10)
print(len(df['html']))
```

```python
df.loc[0, 'is_article'] = 0
df.loc[163, 'is_article'] = 0
df.loc[187:239, 'is_article'] = 0
```

```python
for i, item in enumerate(df.iterrows()):
    soup = BeautifulSoup(item[1].html, 'html.parser')
    res = soup.find(class_="container__link--type-live-story")
    if df.loc[i, 'is_article'] == 0 and res:
        df.loc[i, 'is_article'] = 1
        
len(df[df['is_article']==1])
```
