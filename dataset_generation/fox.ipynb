```python
import sys
sys.path.append("../modules/")
import scraper
from bs4 import BeautifulSoup

html = scraper.fetch_html("https://www.foxnews.com/")
soup = BeautifulSoup(html, 'html.parser')
soup = scraper.clean_html(soup)
soup = scraper.squash_nested_divs(soup)
tags = scraper.extract_tags(soup)

for i, tag in enumerate(tags):
    print(f'{i}: {tag}\n')
```

```python
import pandas as pd

df = pd.DataFrame(tags)
df.columns = ['html']
df['is_article'] = 0
df['i'] = df.index
df.head(10)
print(len(df['html']))
```

```python
for i, item in enumerate(df.iterrows()):
    soup = BeautifulSoup(item[1].html, 'html.parser')
    root = soup.contents[0]
    if (df.loc[i, 'is_article'] == 0 
        and root
        and root.has_attr('class') 
        and 'article' in soup.contents[0]['class']):
        
        soup2 = BeautifulSoup(str(root), 'html.parser')
        a_children = soup2.find_all(['a'])
        for child in a_children:
            df.loc[df.html == str(child), 'is_article'] = 2
        
#     # res = soup.find(class_="gzOvMy")
# #     if df.loc[i, 'is_article'] == 0 and res:
        df.loc[i, 'is_article'] = 1
        
len(df[df['is_article']==1])
```

```python
for i, item in enumerate(df.iterrows()):
    soup = BeautifulSoup(item[1].html, 'html.parser')
    res = soup.find(class_="container__link--type-live-story")
    if df.loc[i, 'is_article'] == 0 and res:
        df.loc[i, 'is_article'] = 1
        
len(df[df['is_article']==1])
```

```python
total = 0
for i, item in enumerate(df.iterrows()):
    if df.loc[i, 'is_article'] == 0:
        total+=1
        print(f"{item[1]['i']}: {item[1]['html']}\n")
        
print(total)
```
