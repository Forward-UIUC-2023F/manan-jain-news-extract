```python
import sys
sys.path.append("../modules/")
import scraper
from bs4 import BeautifulSoup

html = scraper.fetch_html("https://www.ndtv.com/", waitUntil='static')
soup = BeautifulSoup(html, 'html.parser')
soup = scraper.clean_html(soup)
soup = scraper.squash_nested_divs(soup)
tags = scraper.extract_tags(soup)

for i, tag in enumerate(tags):
    print(f'{i}: {tag}\n')
```

```python
import pandas as pd

df = pd.DataFrame(tags)
df.columns = ['html']
df['is_article'] = 0
df['i'] = df.index
df.head(10)
print(len(df['html']))
```

```python
for i, item in enumerate(df.iterrows()):
    soup = BeautifulSoup(item[1].html, 'html.parser')
    root = soup.contents[0]
    if df.loc[i, 'is_article'] == 0 and root:
        soup2 = BeautifulSoup(str(root), 'html.parser')
        img_children = soup2.find_all(['img'])
        if len(img_children) and img_children[0].has_attr('title'):
            df.loc[i, 'is_article'] = 1
        
len(df[df['is_article']==1])
```
